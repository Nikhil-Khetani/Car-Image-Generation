```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv('annotations.csv')
data.head()
```

```python
classes = pd.read_csv('class_names.csv')
classes.head()
```

```python
df = data.join(classes, on='Class', lsuffix='original')
```

```python
df.head()
```

```python
sns.countplot(df['Name'])
```

```python
counts = pd.DataFrame(df['Name'].value_counts())
```

```python
counts.head()
```

```python
counts.reset_index(level=0, inplace=True)
```

```python
counts.tail()
```

```python
sns.histplot(counts)
```

```python
myPlot = sns.barplot(counts['index'],counts['Name'])
myPlot.set(xticklabels=[])
myPlot.set(xlabel='Car Models')
myPlot.set(ylabel = 'Count')
```
